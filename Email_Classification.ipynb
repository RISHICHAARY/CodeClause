# EMAIL SPAM CLASSIFICATION

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("Emails.csv")

print(df.describe())

train_x,test_x,train_y,test_y = train_test_split(df.iloc[:,1:3001],df.iloc[:,-1],test_size = 0.25)
```

```python
svc = SVC(C=1.0,kernel='rbf',gamma='auto')
svc.fit(train_x,train_y)
y_pred = svc.predict(test_x)
print("Accuracy Score for SVC : ", accuracy_score(y_pred,test_y))
```

```python
plt.scatter(train_x.iloc[:,1],train_y,c='r')
plt.scatter(test_x.iloc[:,1],y_pred,c='b')
plt.legend(["Original" , "Predicted"])
plt.show()
```
